## Calculating the Needleman-Wunsch alignment scores with respect to chosen sequence

## sequential execution - bad idea

```python
algns_all = {}
for idx in range(sa.Combined.shape[0]):
    print("At: ", idx)
    algns = sa.calculate_all_alignments(idx)
    algns_all[idx] = algns
```

## parallel execution - better

```python
def func(x):
    print("Parallel at: ", x)
    algns = sa.calculate_all_alignments(x)
    aa = AlignmentAnalyzer(algns)
    scores = aa.get_sorted_scores()
    return (x, scores)
```

```python
n_proc = 16
algn_indices = list(range(sa.Combined.shape[0]))
pool = Pool(processes=n_proc)
res = pool.map(func, algn_indices)
pool.close()
```

```python
len(res)
```

```python
tmptmp = res[0][1]
tmptmp.columns = [0]
tmptmp
```

```python
res[0]
```

```python
similarity_matrix = pd.DataFrame(index=sa.Combined.index)
for col in res:
    tmptmp = col[1]
    tmptmp.columns = [col[0]]
    similarity_matrix = pd.concat([similarity_matrix, tmptmp], axis=1)
    
```

```python
similarity_matrix = pd.concat([similarity_matrix, sa.Combined], axis=1)
similarity_matrix
```

```python
max_scores = []
for col_idx in range(similarity_matrix.shape[0]):
    col = similarity_matrix[col_idx]
    max_item = col.loc[col_idx]
    max_scores.append(max_item)
    similarity_matrix[col_idx] = similarity_matrix[col_idx] / max_item
```

```python
similarity_matrix["max_score"] = max_scores
```
